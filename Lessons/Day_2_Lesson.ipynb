**Sets, Dicts, and Tuples**

Sets are like lists, but cannot contain duplicates. This is useful for when you may want to search for unique items coming from a data set. They are represented with curly braces "{}".

```python
mySet = {1, "a", 5.9, "banana"} #This set contains all kinds of data

print(5 in mySet) # Does the set contain the value 5?
```

```python
mySecondSet = {"a", 8.3, 5.9}

diff = mySet.difference(mySecondSet) # Notice that 8.3 does not appear here..
print(diff)

# What do you think would display if you took mySecondSet and diffed with mySet?

combined = mySet.union(mySecondSet)
print(combined)
```

Dicts, or dictionaries, are like lists with key-value pairs. For example, student ID's and associated names are a fine example of data which may be represented by this format.

```python
my_dict = {
  4380 : "John",
  501 : "Admin",
  4932 : "Bobby"
}

# Who is student 4932?
print(my_dict[4932])
```

```python
# Keys can be any data type. Values can be anything - even lists!

drink_sizes = {
    "giant" : 48,
    "big" : 32,
    "medium" : 24,
    "small" : 18,
    "teeny" : 8,
    "diet" : "empty",
    "combo" : [24, 32, 32, 48]
}

print("{0} {1} {2}".format("Hi can i get a", drink_sizes["giant"], "ounce drink?"))

print(' '.join(("I just drank", str(drink_sizes["medium"] * 4), "oz of Panera lemonade!")))

print(drink_sizes["combo"][2]) # indexing the dict, then indexing the list!
```

Tuples are a whole different beast. Lists are long, ordered groups of objects. Sets and dictionaries are similar with their own spins. Tuples are a python specialty. They're just two things crammed together into one thing. The data types don't matter, and as a result of this, they're not readily available to work with without some fancy syntax. Take a list of coordinates on a map for example. How would you store this in python?

```python
coords = [(103.22, 105.91), (107.49, 105.03), (106.5, 106.88)]
print(type(coords[0]))

# Here, our tuples being stored in this list are just two numbers, grouped
# But what if we want to work with JUST latitudes?

latitudes = list()

for lat, lon in coords: # this splits up the tuple for every tuple in the list
  latitudes.append(lat)

print(latitudes)
```

What if we want to change the latitude for one of our points here? Unfortunately tuples do not support such an operation. You'll have to unpack it and repack it. This makes tuples an "immutable" data type. In essence, you shouldn't ever put in a tuple something which you expect to frequently change. It's still very useful when you have structured pieces of data which need to be treated as one unit, like coordinates.

```python
# Packing a tuple
my_tuple = ("banana", "minion", "baby yoda", "the death penalty")
print(my_tuple[3])

#unpacking the tuple
a, b, c, d = my_tuple
print(b)

new_tuple = (a, b, c, 1984) # replacing one of the original values: a NEW tuple
print(new_tuple)
```

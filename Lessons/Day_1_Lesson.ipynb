**If Statements**

Control flow is the technical term for things like if statements and loops. These statements help to organize the logic of what your program is trying to do. In python, conditionals like the "if statement" are written as follows.

```python
x = 5
if x == 5:
  print("Gravity works")
if x != 5:
  print("Something has gone horribly wrong")

x = x * 3
if x > 10:
  print("And so does math!")
  print("Phew...")
else:
  print("Go home python, you're drunk")

# Comparison operators like > are single characters, but a single = is used
# only for declaring (as in the case of "x = 5"). "==" is used for comparison
```

The indentation here is essential. Python requires that in order for our if statement to have the right "scope", that all internal lines are indented to the same "block". According to PEP8, the official style guide for python, programs should use two spaces in order to differentiate levels of their programs, and never the "Tab" character. Luckily, this notebook is friendly enough to automatically translate any tab characters into the right amount of spaces. In short, don't forget to indent properly when making conditionals, or nested structures like ifs within ifs, or loops within loops.

**Strings**

Strings are pretty special in python. We can actually use indexing to grab specific letters from a string, create substrings from it, reverse the string, etc. The following few lines will demonstrate the power of python strings.

```python
myStr = "War is bad, mkay?"

firstChar = myStr[0] # The first index of any list in software is position 0
print(firstChar)
```

```python
mySubstring = myStr[4:6] # The colon here helps tell python we want a range
print(mySubstring)
```

```python
myNewSubstring = myStr[:10] # Any index left blank assumes we want last/first
print(myNewSubstring)
```

The index after the second colon is stride. Stride tells us "By how many units do we progress through the string?" For example, starting at index 0 and going -1 units each time means we start at index 0, then pull index -1 as our first character. This translates to the last character, which then moves on to the second to last, and so on and so forth.

```python
myBackwardsSubstring = myStr[::-1]
print(myBackwardsSubstring)
```

```python
# print() is pretty friendly in python too and will allow us to do shenanigans
print("We " + mySubstring + " so back" + myStr[-1])
```

**Lists and Loops**

Lists in python are collections of variables which may be of the same or different data types. These lists are ordered, and therefore can be indexed just as we have been doing. Though you didnt realize it, these string slicing exercises have all technically been lists the whole time! Lists are treated as their own data type in python just as "int" and "bool". Lists are declared and invoked with square brackets.

```python
myList = [1, 2, 3]
print(myList[1] + myList[2])

myList = ["hi", "3", True, 4]
print(myList)
```

```python
secondList = [4, 5, 6]
myList.extend(secondList)
print(myList)
```

Lists go hand in hand with loops - which helps to access and manipulate items within the list procedurally. Let's go ahead and take a look at how this works.

```python
myList = [1, 5, 13, 4, 5, 10]

for i in myList:
  print(i + 1)
```

```python
myList = ["a", "b", "c", "d", "e", "f", "g", "h", "i", "j"]

for i in range(0, 5):
  if(myList[i] > "c"):
    print(myList[i])

#The "range()" function is often used in conjunction with the "for" loop.
```
